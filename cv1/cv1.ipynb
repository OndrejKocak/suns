```python
import pandas as pd
import matplotlib as plt
datasetPath = "datasets/IRIS.csv"
iris_df = pd.read_csv(datasetPath)

iris_df.head(50)
```

```python
nullVals = iris_df.isnull().sum().sum()
duplicates = iris_df.duplicated().sum()
samples = iris_df.shape[0]


print(nullVals, duplicates, samples)

iris_df.boxplot(column=['sepal_length','sepal_width', 'petal_length', 'petal_width'])
```

```python
def detectOutliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

column = 'sepal_length'

outliers = detectOutliers(iris_df, column)
plt.figure()
```
